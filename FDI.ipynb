## **All FDI data based on Years**

```python

import altair as alt
import pandas as pd
dataframe_long = pd.melt(dataframe, id_vars=['Sector'], var_name='Year', value_name='Value')

chart = alt.Chart(dataframe_long).mark_line().encode(
    x='Year:O',
    y='Value:Q',
    color='Sector:N'
).properties(
    width=800,
    height=600
)

chart
```

### **FDI in various sectors wise**

> using graphs



```python
total_fdi_by_sector = dataframe_long.groupby('Sector')['Value'].sum().reset_index()

bar_chart = alt.Chart(total_fdi_by_sector).mark_bar().encode(
    x='Sector:N',
    y='Value:Q'
).properties(
    width=800,
    height=600
)

# Display the bar chart
bar_chart
```

## **TOP FIVE SECTORS**

```python
top_5_sectors = total_fdi_by_sector.sort_values('Value', ascending=False).head(5)

print(top_5_sectors)
```

## **Annual growth rate by sector**

```python
dataframe_long['GrowthRate'] = dataframe_long.groupby('Sector')['Value'].pct_change() * 100

dataframe_long['GrowthRate'].replace([np.inf, -np.inf], np.nan, inplace=True)

dataframe_long['GrowthRate'].fillna(0, inplace=True)

growth_chart = alt.Chart(dataframe_long).mark_line().encode(
    x='Year:O',
    y='GrowthRate:Q',
    color='Sector:N'
).properties(
    width=800,
    height=600,
    title='Annual Growth Rate of FDI by Sector'
)

growth_chart
```

## **sector distribution over the years**

*   USING BAR GRAPH



```python
total_fdi_by_year = dataframe_long.groupby('Year')['Value'].sum().reset_index()

stacked_bar_chart = alt.Chart(dataframe_long).mark_bar().encode(
    x='Year:O',
    y='Value:Q',
    color='Sector:N'
).properties(
    width=800,
    height=600,
    title='Sector Distribution of FDI Over the Years'
)
stacked_bar_chart
```

### **Least Performing sectors**

```python
mean_fdi_by_sector = dataframe_long.groupby('Sector')['Value'].mean().reset_index()

bottom_5_sectors = mean_fdi_by_sector.sort_values('Value').head(5)

print(bottom_5_sectors)
```
